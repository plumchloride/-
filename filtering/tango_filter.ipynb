```python
import pandas as pd
```

```python
df = pd.read_csv("./lex_3_1.csv",header=None)
```

```python
# 名詞のみにフィルター
fil_df = df[df[4] == "名詞"]
print(f"{len(df)} => {len(fil_df)}")
```

```python
# 文字数ごとの単語数取得
len_list = []

for i in fil_df[10]:
  len_list.append(len(i))

for i in range(1,40):
  print(f"{i}：{len_list.count(i)}個")
```

```python
# 文字数の列追加（コード非推奨）
fil_df["title_len"] = 5
for index, row in fil_df.iterrows():
  num = len(row[10])
  fil_df["title_len"][index] = int(num)
```

```python
# 指定した単語が含まれているか確認
count = 0
for i in list(fil_df[10]):
  if "ヂ" in i:
    if count < 10:
      print(i)
    count += 1

print(count)
```

```python
# 名詞全部におけるグルーピング
fil_df.groupby([5,6,7]).count()
```

```python
# 5文字の名詞におけるグルーピング
fil_df[fil_df["title_len"]==5].groupby([5,6,7]).count()
```

```python
# 5文字かつ重複はじいた形
fill_5_df = fil_df[fil_df["title_len"]==5]
fill_5_df[~(fill_5_df.duplicated(keep='last',subset=10))].groupby([5,6,7]).count()
```

```python
# 質問用のデータベース取得
Q_df = fil_df[(fil_df[5]=="普通名詞") | ((fil_df[5]=="固有名詞")&(fil_df[6] == "地名")&(fil_df[7] == "国"))]
Q_df = Q_df[Q_df["title_len"] == 5]
Q_df = Q_df[~(Q_df.duplicated(keep='last',subset=10))] # やや不服だが重複を最後だけ残して消す形に
Q_df
```

```python
# 入力用のデータベース取得
A_df = fil_df[(fil_df[5]=="普通名詞") | (fil_df[5]=="固有名詞")]
A_df = A_df[~(A_df[6]=="人名")]
A_df = A_df[A_df["title_len"] == 5]
A_df = A_df[~(A_df.duplicated(keep='last',subset=10))] # やや不服だが重複を最後だけ残して消す形に
A_df
```

```python
# 質問用 csvファイル作成 見出し語（title）と読みのセット
Q_csv_set = Q_df[[0,10]]
Q_csv_set.to_csv(path_or_buf="./Q_data.csv",encoding = "utf-8",header=False, index=False)
```

```python
# 検証用 csvファイル作成 読みのみ
A_csv_set = A_df[10]
A_csv_set.to_csv(path_or_buf="./A_data.csv",encoding = "utf-8",header=False, index=False)
```
