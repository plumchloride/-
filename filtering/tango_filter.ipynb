```python
import pandas as pd
```

```python
df = pd.read_csv("./lex_3_1.csv",header=None)
```

```python
# 名詞のみにフィルター
fil_df = df[df[4] == "名詞"]
print(f"{len(df)} => {len(fil_df)}")
```

```python
# 文字数ごとの単語数取得
len_list = []

for i in fil_df[10]:
  len_list.append(len(i))

for i in range(1,40):
  print(f"{i}：{len_list.count(i)}個")
```

```python
# 文字数の列追加（コード非推奨）
fil_df["title_len"] = 5
for index, row in fil_df.iterrows():
  num = len(row[10])
  fil_df["title_len"][index] = int(num)
```

```python
# 指定した単語が含まれているか確認
count = 0
for i in list(fil_df[10]):
  if "ヂ" in i:
    if count < 10:
      print(i)
    count += 1

print(count)
```

```python
# 名詞全部におけるグルーピング
fil_df.groupby([5,6,7]).count()
```

```python
# 5文字の名詞におけるグルーピング
fil_df[fil_df["title_len"]==5].groupby([5,6,7]).count()
```

```python
# 5文字かつ重複はじいた形
fill_5_df = fil_df[fil_df["title_len"]==5]
fill_5_df[~(fill_5_df.duplicated(keep='last',subset=10))].groupby([5,6,7]).count()
```

```python
# 質問用のデータベース取得
Q_df = fil_df[(fil_df[5]=="普通名詞") | ((fil_df[5]=="固有名詞")&(fil_df[6] == "地名")&(fil_df[7] == "国"))]
Q_df = Q_df[Q_df["title_len"] == 5]
Q_df = Q_df[~(Q_df.duplicated(keep='last',subset=10))] # やや不服だが重複を最後だけ残して消す形に
Q_df
```

```python
# 入力用のデータベース取得
A_df = fil_df[(fil_df[5]=="普通名詞") | (fil_df[5]=="固有名詞")]
A_df = A_df[A_df["title_len"] == 5]
A_df = A_df[~(A_df.duplicated(keep='last',subset=10))] # やや不服だが重複を最後だけ残して消す形に
A_df
```

```python
# 分類語彙表取得
colmuns = ["レコードＩＤ番号","見出し番号","レコード種別","類","部門","中項目","分類項目","分類番号","段落番号","小段落番号","語番号","見出し","見出し本体","読み","逆読み"]
bunrui_df = pd.read_csv("https://raw.githubusercontent.com/masayu-a/WLSP/master/bunruidb.txt",encoding="shiftjis",names=colmuns)
```

```python
bunrui_fil = bunrui_df[bunrui_df["類"]=="体"]
```

```python
bunrui_fil["title_len"] = 3
for index, row in bunrui_fil.iterrows():
  num = len(row[13])
  bunrui_fil["title_len"][index] = int(num)
```

```python
bunrui_5_fil = bunrui_fil[bunrui_fil["title_len"]==5]
bunrui_5_fil
```

```python
# pip install wakana-python
import wanakana
wanakana.to_katakana('あア亜')
bunrui_5_fil["kanayomi"] = "3"
for index, row in bunrui_5_fil.iterrows():
  num = wanakana.to_katakana(row[13])
  bunrui_5_fil["kanayomi"][index] = num
```

```python
tf_list = []
untf_list = []
for i in bunrui_5_fil["見出し本体"]:
  tf_list.append("ー" in i)
  untf_list.append(not("ー" in i))
long_df = bunrui_5_fil[tf_list]
len(long_df)
```

```python
tf_5_list = []
un_tf_5_list = []
for i in long_df["見出し本体"]:
  tf_5_list.append(len(i)!=5)
  un_tf_5_list.append(not(len(i)!=5))
len(long_df[tf_5_list])
```

```python
true_df = long_df[tf_5_list]
syuuseikana = ["ユーターン","ゾウページ","ハートガタ","ヤールハバ","ダーツブン","マレージン","ボーアジン","ダイスター","キーキョク","ブイリーグ","ジスマーク","テキマーク","イーメール","リーグセン","プールセイ","ビールケン","ボールシ","ダンボール","カミテープ","ナマテープ","レースイト","キンモール","ベビーフク","アメコート","マエヨーク","ベレーボウ","スキーボウ","ソーダスイ","ナマビール","クロビール","カンビール","ビールビン","ボールバコ","スキーイタ","アークトウ","ボギーシャ","ループセン","ムールガイ"]
true_df["kanayomi"] = syuuseikana
true_df
```

```python
# 五文字の場合長音が2つより多く入ることはないため、右左からfindすれば置き換えられる
not_diff = long_df[un_tf_5_list]
for index, row in not_diff.iterrows():
  _li = list(row[16])
  _li[row[12].find("ー")] = "ー"
  not_diff["kanayomi"][index] = "".join(_li)

tf_list_2 = []
untf_list_2 = []
for i,z in zip(not_diff["kanayomi"],not_diff["見出し本体"]):
  tf_list_2.append(i.count("ー") != z.count("ー"))
  untf_list_2.append(not(i.count("ー") != z.count("ー")))

print(not_diff[tf_list_2])

for index, row in not_diff[tf_list_2].iterrows():
  _li = list(row[16])
  _li[row[12].rfind("ー")] = "ー"
  not_diff["kanayomi"][index] = "".join(_li)

tf_list_2 = []
untf_list_2 = []
for i,z in zip(not_diff["kanayomi"],not_diff["見出し本体"]):
  tf_list_2.append(i.count("ー") != z.count("ー"))
  untf_list_2.append(not(i.count("ー") != z.count("ー")))

print("--")
print(not_diff[tf_list_2])
not_diff
```

```python
not_long = bunrui_5_fil[untf_list]
Q_bunrui = bunrui_5_fil[["見出し本体","kanayomi"]]
# .to_csv(path_or_buf="./Q_data.csv",encoding = "utf-8",header=False, index=False)
```

```python
# くっつける
Q_bunrui_df = pd.concat([not_long, not_diff,true_df])

# シャッフルする
Q_bunrui_df = Q_bunrui_df.sample(frac=1)
Q_bunrui_df[["見出し本体","kanayomi"]].to_csv(path_or_buf="./Q_data_new.csv",encoding = "utf-8",header=False, index=False)
```
